```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

```python
stock = ["AAPL", "NVDA", "TSLA", "GOOG", "AMZN", "MSFT", "INTC"]

stocks = yf.download(stock, start = "2010-01-01", end = "2024-11-20")
data = stocks.loc[:, "Close"].copy()
```

```python
data.plot(figsize = (17, 8), fontsize = 18)
plt.style.use("seaborn")
plt.show()
```

```python
data.pct_change()
```

```python
com = data.pct_change().dropna()

com.describe()
```

```python
sum = com.describe().T.loc[:, ["mean", "std"]]
sum
```

```python
sum["mean"] = sum["mean"]*252
sum["std"] = sum["std"]*np.sqrt(252)
sum
```

```python
sum.plot.scatter(x = "std", y = "mean", figsize = (17, 8), s = 50, fontsize = 15)
for i in sum.index:
    plt.annotate(i, xy=(sum.loc[i, "std"]+0.002, sum.loc[i, "mean"]+0.002), size = 15)
```
